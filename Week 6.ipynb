```python
#W6A1: RemoveDuplicated
def RemoveDuplicated(a: list) -> list:
    ans = []
    temp = dict()
    for i in a:
        if i not in temp: ans.append(i); temp[i] = 1
    return ans

a = list(map(int, input().split(" ")))
print(RemoveDuplicated(a))
```

```python
#W6A2: PrefixSum
def PrefixSum(a: list) -> list:
    ans = []
    temp = 0
    for i in a: temp += i; ans.append(temp)
    return ans

a = list(map(int, input().split(" ")))
print(PrefixSum(a))
```

```python
#W6A3: k-Left-Handed
def k_left_handed(a: tuple, k: int) -> tuple:
    ans = list()
    left = int(k % len(a))
    for i in range(left, len(a)): ans.append(a[i])
    for i in range(0, left): ans.append(a[i])
    return tuple(ans)

a = list(map(int, input().split()))
k = int(input())
print(k_left_handed(a, k))
```

```python
#W6A4: DictCreator
def DictCreator(a: list) -> dict:
    ans = dict()
    for i in range(0, len(a), 2):
        if a[i] not in ans: ans[a[i]] = list()
        ans[a[i]].append(a[i + 1])
    return ans

a = list(input().split(" "))
print(DictCreator(a))
```

```python
#W6A5: CountSign
def CountSign(a: list) -> dict:
    ans = {
        "positives": 0,
        "negatives": 0,
        "zeros": 0
    }
    for i in a:
        if i > 0: ans["positives"] += 1
        elif i < 0: ans["negatives"] += 1
        else: ans["zeros"] += 1
    return ans
    
a = list(map(int, input().split(" ")))
print(CountSign(a))
```

```python
#W6A9: Concatenate2Dicts
def Concatenate2Dicts(a: dict, b: dict) -> dict:
    for key, val in b.items():
        if key in a: a[key] += val
        else: a[key] = val
    return dict(sorted(a.items()))

m, n = map(int, input().split(" "))
a, b = {}, {}
for i in range(m): k, v = input().split(" "); v = int(v); a[k] = v
for i in range(n): k, v = input().split(" "); v = int(v); b[k] = v
print(a, b)
Concatenate2Dicts(a, b)
```

```python
#W6A10: TwoSumK (coi (a, b) != (b, a))
def TwoSumK(a: list, k: int) -> list:
    repo = {}
    ans = []
    for i in a:
        if i not in repo: repo[i] = 1
        else: repo[i] += 1
    for i in a:
        repo[i] -= 1
        if (k - i in repo) and repo[k - i]: ans.append((i, k - i))
    ans = list(set(ans))
    return sorted(ans, key=lambda x:x[0]) #sorted(ans)??

a = list(map(int, input().split(" ")))
k = int(input())
print(TwoSumK(a, k))
```

```python
#W6A11: Odd/EvenTuples
def Odd_Even_Tuples(a: tuple):
    odd, even = [], []
    for i in a:
        if i % 2: odd.append(i)
        else: even.append(i)
    print(tuple(even), tuple(odd), sep="\n")
    
a = tuple(map(int, input().split(" ")))
Odd_Even_Tuples(a)
```

```python
#W6A12: FindLowestMostElement
def FindLowestMostElement(a: list) -> int:
    cnt = {}
    for i in a:
        if i not in cnt: cnt[i] = 1
        else: cnt[i] += 1
    ans, max_ele = -1, -1
    for key, val in cnt.items():
        if val > max_ele or (val == max_ele and key < ans): ans, max_ele = key, val
    return ans

a = list(map(int, input().split(" ")))
print(FindLowestMostElement(a))
```

```python
#W6A13: Key-Val-Swap
def SwapKeyVal(a: dict) -> dict:
    ans = {}
    for i in a: ans[a[i]] = i
    return ans

n = int(input())
a = {}
for i in range(0, n): v, k = input().split(" "); a[v] = k
if len(a.values()) != len(set(a.values())): print("Value không duy nhất")
else: print(SwapKeyVal(a))
```

```python
#W6A15: FilterDictByVal
def FilterDictByVal(a: dict, k: int) -> dict:
    ans = {}
    for key, v in a.items():
        if v > k: ans[key] = v
    return ans

n = int(input())
a = {}
for i in range(0, n): k, v = input().split(" "); v = int(v); a[k] = v
k = int(input())
print(FilterDictByVal(a, k))
```

```python
#W6B1: ArmstrongNumber
def checkArmstrong(a: int) -> bool:
    total, temp, num_len = 0, a, 0
    while temp:
        num_len += 1
        temp //= 10
    temp = a
    while temp:
        total += (temp % 10) ** num_len
        temp //= 10
    return total == a

a = int(input())
print(checkArmstrong(a))
```

```python
#W6B2: CheckSquare
def CheckSquare(a: int) -> bool:
    temp = 0
    while temp ** 2 < a: temp += 1
    return temp ** 2 == a

a = int(input())
print(CheckSquare(a))
```

```python
#W6B3: CheckPalindromeNum
def CheckPalindromeNum(a: int) -> bool:
    temp = []
    while a:
        temp.append(a % 10)
        a //= 10
    return temp == temp[::-1]

a = int(input())
print(CheckPalindromeNum(a))
```

```python
#W6B4: CheckMagicNum
def CheckMagicNum(a: int) -> bool:
    while a > 9:
        temp = 0
        while a: temp += (a % 10); a //= 10
        a = temp
    return a == 1

a = int(input())
print(CheckMagicNum(a))
```
