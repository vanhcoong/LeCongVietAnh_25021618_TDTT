```python
#W5A1: def Max_Of_Two(a, b) -> int
def Max_Of_Two(a, b):
    return max(a, b)

def main():
    a, b = map(int, input().split(" "))
    print(Max_Of_Two(a, b))
    
if __name__ == "__main__": main()
```

```python
#W5A2: def swap(a, b) -> Tuple[int, int]
def swap(a, b):
    return (b, a)

def main():
    a, b = map(int, input().split(" "))
    print(swap(a, b))

if __name__ == "__main__": main()
```

```python
#W5A3: def CheckPrime(a) -> bool
def CheckPrime(a):
    if a <= 1: return False
    temp = 2
    while temp * temp <= a:
        if a % temp == 0: return False
        temp += 1
    return True

def main():
    a = int(input())
    print(CheckPrime(a))
    
if __name__ == "__main__": main()        
```

```python
#W5A4: def CheckPerfect(int a) -> bool
def CheckPerfect(a):
    total, temp = 1, 2
    if a <= 1: return False
    while temp * temp < a:
        if a % temp == 0: total += (temp + a / temp)
        temp += 1
    if temp * temp == a: total += temp
    return a == total

def main():
    n = int(input())
    print(CheckPerfect(n))
    
if __name__ == "__main__": main()
```

```python
#W5A5: def FirstK(a, k) -> int
def FirstK(a, k):
    for i in range(0, len(a)):
        if a[i] == k: return i + 1
    else: return -1
    
def main():
    a = list(map(int, input().split()))
    k = int(input())
    print(FirstK(a, k))
    
if __name__ == "__main__": main()
```

```python
#W5A6: def Factorial(int a) -> int
def Factorial(a):
    if a < 0: return -1
    ans = 1
    for i in range(1, a + 1): ans *= i
    return ans

def main():
    a = int(input())
    print(Factorial(a))

if __name__ == "__main__": main()
```

```python
#W5A7: def Calculator(float a, float b, string op) -> float (round to 2 decimal digits)
def Calculator(a, b, c):
    match c:
        case '+': return a + b
        case '-': return a - b
        case '*': return a * b
        case '/': return a / b

def main():
    a, b, c = input().split(" ")
    a, b = float(a), float(b)
    print(f'{Calculator(a, b, c):.2f}')

if __name__ == "__main__": main()
```

```python
#W5A8: def Hamming(a, b) -> int
def Hamming(a, b):
    ans = 0
    while a or b:
        if a % 2 != b % 2: ans += 1
        a //= 2
        b //= 2
    return ans

def main():
    a, b = map(int, input().split())
    print(Hamming(a, b))
    
if __name__ == "__main__": main()
```

```python
#W5A9: def SumDigit(a) -> int
def SumDigit(a):
    ans = 0
    while a: ans += a % 10; a //= 10
    return ans

def main():
    a = abs(int(input()))
    print(SumDigit(a))
    
if __name__ == "__main__": main()
```

```python
#W5A10: def Isomorphism(a, b) -> bool
def HalfIsomorphim(a, b):
    temp = [-1 for i in range(0, 26)]
    for i in range(0, len(a)):
        if temp[ord(a[i]) - ord('a')] == -1: temp[ord(a[i]) - ord('a')] = ord(b[i]) - ord('a')
        elif temp[ord(a[i]) - ord('a')] != ord(b[i]) - ord('a'): return False
    return True

def IsoMorphism(a, b): return HalfIsomorphim(a, b) and HalfIsomorphim(b, a)

def main():
    a, b = input().lower().split(" ")
    print(IsoMorphism(a, b))
    
if __name__ == "__main__": main() 
```
