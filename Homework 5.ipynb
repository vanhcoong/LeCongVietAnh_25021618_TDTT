```python
#W6A16: RePrint (Matrix nxm)
def RePrint(a):
    for i in range(0, len(a)):
        for j in range(0, len(a[i])): print(a[i][j], end=" ")
        print()
        
n, m = map(int, input().split(" "))
a = [[] for i in range(0, n)]
for i in range(0, n): a[i] = list(map(int, input().split(" ")))
RePrint(a) 
```

```python
#W6A17: MatrixDiagonal
def MatrixDiagonal(a: list):
    n = len(a)
    for i in range(0, n): print(a[i][i], end=" ")
    print()
    for i in range(0, n): print(a[i][n - 1 - i], end=" ")
    
n = int(input())
a = [[] for i in range(0, n)]
for i in range(0, n): a[i] = list(map(int, input().split(" ")))
MatrixDiagonal(a)
```

```python
#W6A20: FindFirstK
def FindFirstK(a, k):
    for i in range(0, len(a)):
        if a[i] == k: return i + 1
    else: return -1

a = list(map(int, input().split(" ")))
k = int(input())
print(FindFirstK(a, k))
```

```python
#W6A21: DictStrPlaces
def DictStrPlaces(a):
    ans = {}
    cnt = 1
    for i in a:
        if i not in ans: ans[i] = []
        ans[i].append(cnt)
        cnt += 1
    return ans

a = input().split(" ")
print(DictStrPlaces(a))
```

```python
#W6A23: DelLessX
def DelLessX(a):
    x = int(input())
    ans = []
    for i in a:
        if i >= x: ans.append(i)
    return ans

a = list(map(int, input().split(" ")))
print(DelLessX(a))
```

```python
#W6A24: 0-to-the-left
def zero_to_the_left(a):
    num_0 = 0
    ans = []
    for i in a:
        if i: ans.append(i)
        else: num_0 += 1
    return [0] * num_0 + ans

a = list(map(int, input().split(" ")))
print(zero_to_the_left(a))
```

```python
#W6A25: NotFound0_to_9
def NotFound0_to_9(a: list) -> list:
    ans = []
    cnt = {i: 0 for i in range(10)}
    for i in a: cnt[i] += 1
    for i in range(10):
        if cnt[i] == 0: ans.append(i)
    return ans

a = list(map(int, input().split(" ")))
print(NotFound0_to_9(a))
```

```python
#W6A26: SumLocalMaximum (Coi LocalMaximum = {ai|ai > ai-1 và ai > ai+1})
def SumLocalMaximum(a: list) -> int:
    if len(a) <= 2: return 0
    ans = 0
    for i in range(1, len(a) - 1):
        if a[i] > a[i - 1] and a[i] > a[i + 1]: ans += a[i]
    return ans

a = list(map(int, input().split(" ")))
print(SumLocalMaximum(a))
```

```python
#W6A27: FindLessX
def FindLessX(a: list, x: int) -> list:
    ans = []
    for i in range(0, len(a)):
        if a[i] < x: ans.append(i + 1)
    return ans

a = list(map(int, input().split(" ")))
x = int(input())
fin = FindLessX(a, x)
print(fin if len(fin) else -1)
```

```python
#W6A28: AbsoluteIncDec
def AbsoluteIncDec(a):
    if len(a) == 1: return "Vô danh"
    inc = dec = True
    for i in range(1, len(a)):
        if a[i] > a[i - 1]: dec = False
        elif a[i] < a[i - 1]: inc = False
        else: inc = dec = False
        if inc == dec == False: return "Vô danh"
    return "Dãy tăng thực sự" if inc else "Dãy giảm thực sự"

a = list(map(int, input().split(" ")))
print(AbsoluteIncDec(a))
```
